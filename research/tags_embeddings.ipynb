```python
import numpy as np
import pandas as pd
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
```

```python
with open("titles_with_tags_dict.pkl", "rb") as f:
    data = pickle.load(f)
```

```python
df = pd.read_excel("data_clean.xlsx", header=0)
```

```python
df["payload"] = df.apply(lambda x: json.dumps(data.get(x["title_rus"])), axis=1)
```

```python
df.head()
```

```python
corpus_texts = df["payload"].tolist()
```

```python
tags_embeddings = model.encode(corpus_texts, normalize_embeddings=True)
```

```python
with open("tags_embeddings.pkl", "wb") as f:
    pickle.dump(tags_embeddings, f)
```
